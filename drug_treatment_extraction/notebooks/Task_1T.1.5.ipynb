```python
#This notebook uses the virus / therapy overlaps and seeks possible hedge words
#  in the sentences where the match was found.

import numpy as np
import pylab
import pandas as pd
import json
import os
import scispacy
import spacy
import re
import time
nlp=spacy.load("en_core_sci_lg")
```

```python
overlap=pd.read_csv("Overlaps_Virus_Therapy.csv")
```

```python
SentenceEnders="\. |; \! \? "
HedgeWords=['might','potential','could','perhaps','possible','possibly','may','tentative','hypothesize','speculate','speculatively']
Sentences={}
Hedges={}

for i in overlap[overlap.same_sentence==1].index:
    sentences = re.split(SentenceEnders,overlap.block[i])
    # print(sentences[0])
    for s in sentences:
        s=s.lower()
        if((overlap.word_therapy[i] in s) and (overlap.word_virus[i] in s)):
            Sentences[i]=s
         #   print(s)
            for n in HedgeWords:
                if(n in s):
                    Hedges[i]=n
```

```python
overlap.insert(len(overlap.columns),'sentence',np.NaN)
overlap.insert(len(overlap.columns),'hedge',np.NaN)
```

```python
for k in Sentences.keys():
    overlap.sentence[k]=Sentences[k]
for k in Hedges.keys():
    overlap.hedge[k]=Hedges[k]
```

```python
for s in overlap[(overlap.same_sentence==1)& (overlap.notna().hedge)].sentence.unique():
    print(s)
    print( " --> POSSIBLE HEDGE")
    print("")
```
